# Q1

```python
with open("Day 4.txt") as f:
    lines = f.readlines()
    paper_map = {(i, j): c for i, line in enumerate(lines) for j, c in enumerate(line)}

    accessible_rolls = 0

    for (i, j), v in paper_map.items():
        if v == "@":
            accessible_rolls += (
                sum(
                    [
                        c == "@"
                        for c in [
                            paper_map.get((i - 1, j - 1), "."),
                            paper_map.get((i - 1, j), "."),
                            paper_map.get((i - 1, j + 1), "."),
                            paper_map.get((i, j - 1), "."),
                            paper_map.get((i, j + 1), "."),
                            paper_map.get((i + 1, j - 1), "."),
                            paper_map.get((i + 1, j), "."),
                            paper_map.get((i + 1, j + 1), ".")
                        ]
                    ]
                )
                < 4
            )

accessible_rolls
```

# Q2

```python
with open("Day 4.txt") as f:
    lines = f.readlines()
    paper_map = {(i, j): c for i, line in enumerate(lines) for j, c in enumerate(line)}

    to_visit = set(paper_map.keys())
    removed_rolls = 0

    while to_visit:
        i, j = to_visit.pop()

        if paper_map.get((i, j), ".") == "@":
            to_check = {
                (i - 1, j - 1),
                (i - 1, j),
                (i - 1, j + 1),
                (i, j - 1),
                (i, j + 1),
                (i + 1, j - 1),
                (i + 1, j),
                (i + 1, j + 1)
            }

            papers = {(i, j) for i, j in to_check if paper_map.get((i, j), ".") == "@"}

            if len(papers) < 4:
                paper_map[(i, j)] = "."
                removed_rolls += 1
                to_visit = to_visit.union(papers)

removed_rolls
```
