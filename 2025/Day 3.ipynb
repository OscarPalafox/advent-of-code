# Q1

```python
with open("Day 3.txt") as f:
    tot = 0

    while line := f.readline():
        line = line.strip()

        bat1 = line[0]
        bat2 = line[1]

        for i in range(2, len(line)):
            n = line[i]
            if bat2 > bat1:
                bat1 = bat2
                bat2 = n
            elif n > bat2:
                bat2 = n

        tot += int(bat1 + bat2)

tot
```

# Q2

```python
with open("Day 3.txt") as f:
    tot = 0

    while line := f.readline():
        line = line.strip()

        bat = list(line[:12])

        for i in range(12, len(line)):
            bat.append(line[i])

            for idx_to_remove, val_to_remove in enumerate(bat[:-1]):
                if val_to_remove < bat[idx_to_remove + 1]:
                    break
            else:
                idx_to_remove += 1

            bat.pop(idx_to_remove)

        tot += int("".join(bat))

tot
```
