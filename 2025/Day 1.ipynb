# Q1

```python
with open("Day 1.txt") as f:
    pos = 50
    res = 0

    while line := f.readline():
        line = line.strip()

        mul = {
            "L": -1,
            "R": 1
        }

        pos = (pos + (mul[line[0]] * int(line[1:]))) % 100

        if not pos:
            res += 1

res
```

# Q2

```python
with open("Day 1.txt") as f:
    pos = 50
    res = 0

    while line := f.readline():
        line = line.strip()

        mul = {"L": -1, "R": 1}

        prev_pos = pos
        pos = (pos + (mul[line[0]] * int(line[1:]))) % 100

        res += int(line[1:]) // 100
        if not pos:
            res += 1
        elif (
            prev_pos
            and (
                (pos > prev_pos and mul[line[0]] < 0)
                or (pos < prev_pos and mul[line[0]] > 0)
            )
        ):
            res += 1

res
```
