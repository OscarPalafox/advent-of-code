# Q1

```python
import math

with open("Day 2.txt") as f:
    ranges = f.readline().strip().split(",")

    tot = 0

    for r in ranges:
        s, e = [int(i) for i in r.split("-")]

        digs = int(math.log10(s))

        if (e <= 10) or ((digs == int(math.log10(e))) and not (digs % 2)):
            continue

        if not (digs % 2):
            s = 10**(digs + 1)

        l = int(str(s)[:digs // 2 + 1])

        if int(str(s)[digs // 2 + 1:]) > l:
            l += 1

        r = l

        while (curr := r + l * 10**int(math.log10(r) + 1)) <= e:
            tot += curr

            l += 1
            r += 1


tot
```

# Q2

```python
with open("Day 2.txt") as f:
    ranges = f.readline().strip().split(",")

    tot = 0

    for r in ranges:
        s, e = [int(i) for i in r.split("-")]

        for n in range(s, e + 1):
            num = str(n)

            for i in range(1, len(num)):
                if not len(num) % i and num[:i] * (len(num) // i) == num:
                    tot += n
                    break

tot
```
