# Task 2

```python
import numpy as np

with open("day 8.txt") as f:
    lines = f.readlines()

lines = lines
inputs = [line.rstrip().split(" | ")[0].split(" ") for line in lines]
outputs = [line.rstrip().split(" | ")[1].split(" ") for line in lines]

def fill_board(arr_input):
    board = np.zeros(7*6, dtype="O").reshape((7, 6))
    vec_len = np.vectorize(len)

    ones = arr_input[vec_len(arr_input) == 2]
    if ones:
        board[1:3, 5] = ones[0]
        board[4:6, 5] = ones[0]
    
    sevens = arr_input[vec_len(arr_input) == 3]
    if sevens:
        if ones:
            top = (set(sevens[0]) - set(ones[0])).pop()
            board[0, 1:5] = top
        else:
            board[1:3, 5] = sevens[0]
            board[4:6, 5] = sevens[0]
            board[0, 1:5] = sevens[0]

    fours = arr_input[vec_len(arr_input) == 4]
    if fours:
        if ones:
            left = "".join(set(fours[0]) - set(ones[0]))
            board[1:3, 0] = left
            board[3, 1:5] = left
        else:
            board[1:3, 5] = fours[0]
            board[4:6, 5] = fours[0]
            board[1:3, 0] = fours[0]
            board[3, 1:5] = fours[0]

    eights = arr_input[vec_len(arr_input) == 7]
    if eights:
        curr_info = (set(board[0, 1]).union(set(board[1, 0]))
                                     .union(set(board[3, 1]))
                                     .union(set(board[4, 5])))
        left = "".join(set(eights[0]) - curr_info)
        board[4:6, 0:1] = left
        board[6, 1:5] = left

    mid_row = board[3, 1]
    if mid_row:
        poss_zeros = arr_input[vec_len(arr_input) == 6]
        poss_zeros0 = poss_zeros[np.where(np.char.find(poss_zeros, mid_row[0]) != -1)]
        poss_zeros1 = poss_zeros[np.where(np.char.find(poss_zeros, mid_row[1]) != -1)]

        if len(poss_zeros0) > len(poss_zeros1):
            zeros = list(set(poss_zeros0) - set(poss_zeros1))
        elif len(poss_zeros1) > len(poss_zeros0):
            zeros = list(set(poss_zeros1) - set(poss_zeros0))
        else:
            zeros = []
        
        if zeros:
            top_left = set(mid_row).intersection(set(zeros[0])).pop()
            board[1:3, 0] = top_left

            mid_row = (set(mid_row) - set(top_left)).pop()
            board[3, 1:5] = mid_row
    
    bot_row = board[6, 1]
    if bot_row:
        poss_nines = arr_input[vec_len(arr_input) == 6]
        poss_nines0 = poss_nines[np.where(np.char.find(poss_nines, bot_row[0]) != -1)]
        poss_nines1 = poss_nines[np.where(np.char.find(poss_nines, bot_row[1]) != -1)]

        if len(poss_nines0) > len(poss_nines1):
            nines = list(set(poss_nines0) - set(poss_nines1))
        elif len(poss_nines1) > len(poss_nines0):
            nines = list(set(poss_nines1) - set(poss_nines0))
        else:
            nines = []
        
        if nines:
            bot_row = set(bot_row).intersection(set(nines[0])).pop()
            board[6, 1:5] = bot_row
            
            bot_left = (set(board[4, 0]) - set(bot_row)).pop()
            board[4:6, 0] = bot_left
    
    top_right = board[1, 5]
    if top_right:
        poss_fives = arr_input[vec_len(arr_input) == 5]
        poss_fives0 = poss_fives[np.where(np.char.find(poss_fives, top_right[0]) != -1)]
        poss_fives1 = poss_fives[np.where(np.char.find(poss_fives, top_right[1]) != -1)]

        if len(poss_fives0) > len(poss_fives1):
            fives = list(set(poss_fives0) - set(poss_fives1))
        elif len(poss_fives1) > len(poss_fives0):
            fives = list(set(poss_fives1) - set(poss_fives0))
        else:
            fives = []

        if fives:
            bot_right = set(top_right).intersection(set(fives[0])).pop()
            board[4:6, 5] = bot_right

            top_right = (set(top_right) - set(bot_right)).pop()
            board[1:3, 5] = top_right
            
        else:
            poss_sixes = arr_input[vec_len(arr_input) == 6]
            poss_sixes0 = poss_sixes[np.where(np.char.find(poss_sixes, top_right[0]) != -1)]
            poss_sixes1 = poss_sixes[np.where(np.char.find(poss_sixes, top_right[1]) != -1)]

            if len(poss_sixes0) > len(poss_sixes1):
                sixes = list(set(poss_sixes0) - set(poss_sixes1))
            elif len(poss_sixes1) > len(poss_sixes0):
                sixes = list(set(poss_sixes1) - set(poss_sixes0))
            else:
                sixes = []

            if sixes:
                bot_right = set(top_right).intersection(set(sixes[0])).pop()
                board[4:6, 5] = bot_right

                top_right = (set(top_right) - set(bot_right)).pop()
                board[1:3, 5] = top_right


    return board


out_values = []

for i, output in enumerate(outputs):
    board = fill_board(np.array(inputs[i]))

    let_num = {
        frozenset([board[0, 1], board[1, 0], board[1, 5],
                   board[4, 0], board[4, 5], board[6, 1]]): "0",
        frozenset([board[1, 5], board[4, 5]]): "1",
        frozenset([board[0, 1], board[1, 5], board[3, 1],
                   board[4, 0], board[6, 1]]): "2",
        frozenset([board[0, 1], board[1, 5], board[3, 1],
                   board[4, 5], board[6, 1]]): "3",
        frozenset([board[1, 0], board[1, 5], board[3, 1],
                   board[4, 5]]): "4",
        frozenset([board[0, 1], board[1, 0], board[3, 1],
                   board[4, 5], board[6, 1]]): "5",
        frozenset([board[0, 1], board[1, 0], board[3, 1],
                   board[4, 0], board[4, 5], board[6, 1]]): "6",
        frozenset([board[0, 1], board[1, 5], board[4, 5]]): "7",
        frozenset([board[0, 1], board[1, 0], board[1, 5],
                   board[3, 1], board[4, 0], board[4, 5],
                   board[6, 1]]): "8",
        frozenset([board[0, 1], board[1, 0], board[1, 5],
                   board[3, 1], board[4, 5], board[6, 1]]): "9",
    }
    output_digit_num = ""

    for output_digit in output:
        output_digit_num += let_num[frozenset(output_digit)]
    
    out_values.append(int(output_digit_num))

sum(out_values)
```
