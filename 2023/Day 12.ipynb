```python
spring = "????##?"
spring_next = f"?{spring}"
group_sizes = [1, 1, 3]

arrangements = 0
rep_idx = [idx for idx, char in enumerate(spring) if char == "?"]
possibilities = list(itertools.product(["#", "."], repeat=len(rep_idx)))
# Handle unfolded
rep_idx_next = [idx for idx, char in enumerate(spring_next) if char == "?"]
possibilities_next = list(itertools.product(["#", "."], repeat=len(rep_idx_next)))

springs = replace_chars_spring(spring, rep_idx, possibilities)
springs_next = replace_chars_spring(spring_next, rep_idx_next, possibilities_next)

group_sizes_to_spring = collections.defaultdict(set)
group_sizes_to_spring_next = collections.defaultdict(set)

for spring_poss in springs:
    if valid_poss_start(group_sizes, extract_sizes(spring_poss)):
        group_sizes_to_spring[tuple(extract_sizes(spring_poss))].add(spring_poss)

for spring_poss_next in springs_next:
    if valid_poss(group_sizes, extract_sizes(spring_poss_next)):
        group_sizes_to_spring_next[tuple(extract_sizes(spring_poss_next))].add(spring_poss_next)

tot_combs = 0
for first_condition, combs in group_sizes_to_spring.items():
    for sec_condition, sec_combs in group_sizes_to_spring_next.items():
        tot_len = len(first_condition)

        if sec_condition[0] != group_sizes[tot_len % len(group_sizes)]:
            continue

        for thi_condition, thi_combs in group_sizes_to_spring_next.items():
            tot_len = len(first_condition) + len(sec_condition)

            if thi_condition[0] != group_sizes[tot_len % len(group_sizes)]:
                continue

            for fou_condition, fou_combs in group_sizes_to_spring_next.items():
                tot_len = len(first_condition) + len(sec_condition) + len(thi_condition)

                if fou_condition[0] != group_sizes[tot_len % len(group_sizes)]:
                    continue

                for fiv_condition, fiv_combs in group_sizes_to_spring_next.items():
                    tot_len = len(first_condition) + len(sec_condition) + len(thi_condition) + len(fou_condition)

                    if fiv_condition[0] != group_sizes[tot_len % len(group_sizes)]:
                        continue

                    if tot_len + len(fiv_condition) == len(group_sizes) * 5:
                        x = len(combs) * len(sec_combs) * len(thi_combs) * len(fou_combs) * len(fiv_combs)
                        tot_combs += len(combs) * len(sec_combs) * len(thi_combs) * len(fou_combs) * len(fiv_combs)

group_sizes_to_spring, group_sizes_to_spring_next
```
