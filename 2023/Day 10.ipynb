```python
def next_tile(
        prev_pos: tuple[int, int],
        pos: tuple[int, int],
    ) -> tuple[int, int]:
    tile = pipes_mapped[pos]
    if tile == "|":
        return (pos[0] + (pos[0] - prev_pos[0]), pos[1])
    elif tile == "-":
        return (pos[0], pos[1] + (pos[1] - prev_pos[1]))
    elif tile == "L":
        return (pos[0] + (pos[1] - prev_pos[1]), pos[1] + (pos[0] - prev_pos[0]))
    elif tile == "J":
        return (pos[0] - (pos[1] - prev_pos[1]), pos[1] - (pos[0] - prev_pos[0]))
    elif tile == "7":
        return (pos[0] + (pos[1] - prev_pos[1]), pos[1] + (pos[0] - prev_pos[0]))
    else:
        return (pos[0] - (pos[1] - prev_pos[1]), pos[1] - (pos[0] - prev_pos[0]))


def get_start_dirs(start_pos: tuple[int, int]):
    dirs = []

    top = pipes_mapped.get(top_pos := (start_pos[0] - 1, start_pos[1]), ".")
    top_bool = False
    left = pipes_mapped.get(left_pos := (start_pos[0], start_pos[1] - 1), ".")
    left_bool = False
    right = pipes_mapped.get(right_pos := (start_pos[0], start_pos[1] + 1), ".")
    right_bool = False
    bot = pipes_mapped.get(bot_pos := (start_pos[0] + 1, start_pos[1]), ".")
    bot_bool = False

    if top in ["|", "7", "F"]:
        dirs.append(top_pos)
        top_bool = True

    if left in ["-", "L", "F"]:
        dirs.append(left_pos)
        left_bool = True

    if right in ["-", "J", "7"]:
        dirs.append(right_pos)
        right_bool = True

    if bot in ["|", "L", "J"]:
        dirs.append(bot_pos)
        bot_bool = True

    if top_bool and right_bool:
        return dirs, "L"
    elif top_bool and left_bool:
        return dirs, "J"
    elif bot_bool and right_bool:
        return dirs, "F"
    elif bot_bool and left_bool:
        return dirs, "7"
    elif top_bool and bot_bool:
        return dirs, "|"
    else:
        return dirs, "-"

def bfs(row: int, col: int) -> bool:
    visited = set([(row, col)])
    queue = set([(row, col)])

    while queue:
        i, j = queue.pop()
        visited.add((i, j))

        if i in {0, len(raw_map) - 1} or j in {0, len(raw_map[0]) - 1}:
            return False

        # Top
        if raw_map[i - 1][j] in {",", "."}:
            if (top := (i - 1, j)) not in visited:
                queue.add(top)

        # Left
        if raw_map[i][j - 1] in {",", "."}:
            if (left := (i, j - 1)) not in visited:
                queue.add(left)

        # Right
        if raw_map[i][j + 1] in {",", "."}:
            if (right := (i, j + 1)) not in visited:
                queue.add(right)

        # Bot
        if raw_map[i + 1][j] in {",", "."}:
            if (bot := (i + 1, j)) not in visited:
                queue.add(bot)

    return True


with open("Day 10.txt") as f:
    pipes_mapped = {}
    start = (-1, -1)

    line_idx = 0
    raw_map = []
    while line := f.readline():
        raw_map.append(list(line.strip()))
        for i, char in enumerate(line.strip()):
            pipes_mapped[line_idx, i] = char

            if char == "S":
                start = (line_idx, i)

        line_idx += 1

(way_1, way_2), start_replace = get_start_dirs(start)
pos_1 = pos_2 = start
loop = set([start, way_1, way_2])
while way_1 != way_2:
    way_1, pos_1 = next_tile(pos_1, way_1), way_1
    loop.add(way_1)
    way_2, pos_2 = next_tile(pos_2, way_2), way_2
    loop.add(way_2)

for i in range(len(raw_map)):
    for j, tile in enumerate(raw_map[i]):
        if (i, j) not in loop:
            raw_map[i][j] = "."

raw_map[start[0]][start[1]] = start_replace

# Insert "," in horizontal spaces
i = 1
while i < len(raw_map):
    if {"L", "-", "7"}.intersection(set(raw_map[i])):
        raw_map.insert(i, [","] * len(raw_map[i]))

        for j in range(len(raw_map[i])):
            if raw_map[i + 1][j] in {"|", "J", "L"}:
                raw_map[i][j] = "|"
        i += 1
    i += 1

# Insert "," in vertical spaces
j = 1
while j < len(raw_map[0]):
    col = [raw_map[i][j] for i in range(len(raw_map))]
    if {"F", "|", "L"}.intersection(set(col)):
        for i in range(len(raw_map)):
            raw_map[i].insert(j, ",")

        for i in range(len(raw_map)):
            if raw_map[i][j + 1] in {"-", "J", "7"}:
                raw_map[i][j] = "-"
        j += 1
    j += 1

enclosed_tiles = 0
for i in range(len(raw_map)):
    for j, tile in enumerate(raw_map[i]):
        if tile == ".":
            if bfs(i, j):
                enclosed_tiles += 1


enclosed_tiles
```
